```python
import random as rn
import numpy as np
import math as m
import matplotlib.pyplot as plt

#batch gradient descent
def bgd(ytr, xtr, nfeat, maxrun, conv, alpha):
    print "Performing gradient descent"
    neg = np.shape(ytr)[0]
    print neg
    
    w = np.random.rand(1, nfeat+1)
    lse = np.zeros(maxrun)
    
    for r in range(0,maxrun):
        yh = np.transpose(np.dot(w, np.transpose(xtr)))
        diff = np.subtract(yh, ytr)
        gradient = np.sum(np.dot(np.transpose(diff),xtr))/neg
        w = np.subtract(w, alpha*gradient)
        lse[r] = np.sum(np.multiply(diff,diff))/neg
        
        if(r>1):
            if(lse[r]>lse[r-1]):
                print "Diverging; exiting"
                lse = np.array(lse[1:r])
                break
        
        print r, lse[r]
        if lse[r]<conv:
            print "Converged"
            lse = np.array(lse[1:r])
            break
    return w, lse

#stochastic gradient descent
def sgd(ytr, xtr, nfeat, maxrun, conv, alpha):
    print "Performing stochastic gradient descent"
    neg = np.shape(ytr)[0]
    
    w = np.random.rand(1, nfeat+1)
    lse = np.zeros(maxrun)
    
    for r in range(0,maxrun):
        for i in range(0,neg):
            yh = np.dot(w, xtr[i])
            diff = yh-ytr[i]
            gradient = diff*xtr[i]/neg
            w = np.subtract(w, alpha*gradient)
            lse[r] += (diff**2)/neg
        
        if(r>1):
            if(lse[r]>lse[r-1]):
                print "Diverging; exiting"
                lse = np.array(lse[1:r])
                break
        
        print r, lse[r]
        if lse[r]<conv:
            print "Converged"
            lse = np.array(lse[1:r])
            break
            
    return w, lse
        
#stochastic gradient descent with momentum
def sgdm(ytr, xtr, nfeat, maxrun, conv, alpha, momentum):
    print "Performing stochastic gradient descent with momentum"
    neg = np.shape(ytr)[0]
    
    w = np.random.rand(1, nfeat+1)
    delw = np.zeros(nfeat+1)
    lse = np.zeros(maxrun)
    
    for r in range(0,maxrun):
        for i in range(0,neg):
            yh = np.dot(w, xtr[i])
            diff = yh-ytr[i]
            gradient = diff*xtr[i]/neg
            w = np.subtract(w, alpha*gradient)
            w = np.add(w, momentum*delw)
            delw = np.add(-alpha*gradient, momentum*delw)
            lse[r] += (diff**2)/neg
        
        if(r>1):
            if(lse[r]>lse[r-1]):
                print "Diverging; exiting"
                lse = np.array(lse[1:r])
                break
        
        print r, lse[r]
        if lse[r]<conv:
            print "Converged"
            lse = np.array(lse[1:r])
            break
            
    return w, lse

#adagrad
def adagrad(ytr, xtr, nfeat, maxrun, conv, alpha):
    print "Performing AdaGrad"
    neg = np.shape(ytr)[0]
    
    w = np.random.rand(1, nfeat+1)
    g2 = np.zeros(nfeat+1)
    lse = np.zeros(maxrun)

    for r in range(0,maxrun):
        for i in range(0,neg):
            yh = np.dot(w, xtr[i])
            diff = yh-ytr[i]
            gradient = diff*xtr[i]/neg
            g2 = np.add(g2, np.multiply(gradient, gradient))
            w = np.subtract(w, np.divide(alpha*gradient, np.sqrt(g2)))
            lse[r] += (diff**2)/neg
        
        if(r>1):
            if(lse[r]>lse[r-1]):
                print "Diverging; exiting"
                lse = np.array(lse[1:r])
                break
        
        print r, lse[r]
        if lse[r]<conv:
            print "Converged"
            lse = np.array(lse[1:r])
            break
            
    return w, lse

#rmsprop
def rmsprop(ytr, xtr, nfeat, maxrun, conv, alpha, ff):
    print "Performing AdaGrad"
    neg = np.shape(ytr)[0]
    
    w = np.random.rand(1, nfeat+1)
    gavg = np.zeros(nfeat+1)
    lse = np.zeros(maxrun)

    for r in range(0,maxrun):
        for i in range(0,neg):
            yh = np.dot(w, xtr[i])
            diff = yh-ytr[i]
            gradient = diff*xtr[i]/neg
            gavg = np.add(ff*gavg, (1-ff)*np.multiply(gradient, gradient))
            w = np.subtract(w, np.divide(alpha*gradient, np.sqrt(gavg)))
            lse[r] += (diff**2)/neg
        
        if(r>1):
            if(lse[r]>lse[r-1]):
                print "Diverging; exiting"
                lse = np.array(lse[1:r])
                break
        
        print r, lse[r]
        if lse[r]<conv:
            print "Converged"
            lse = np.array(lse[1:r])
            break
            
    return w, lse

#creating dataset
n = 1000;
x = np.zeros(n)
y = np.zeros(n)
amplitude = 0.1
nfeat = 3

x = 10*np.random.rand(n, nfeat)
x = np.insert(x, 0, 1, axis=1)
w_actual = np.random.rand(1, nfeat+1)
y = np.transpose(np.dot(w_actual, np.transpose(x)))

for i in range(0, n):
    check = rn.random()
    if(check < 0.5):
        y[i] += amplitude*rn.random()
        
split = (int)(n*0.7)
ytr = np.array(y[1:split])
yts = np.array(y[split+1:n])
xtr = np.array(x[1:split])
xts = np.array(x[split+1:n])

#w, lse = sgdm(ytr, xtr, nfeat, 100000, 0.1, 1.0e-3, 0.1)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3, Momentum = 0.1")

#w, lse = sgdm(ytr, xtr, nfeat, 100000, 0.1, 1.0e-3, 0.1)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3, Momentum = 1.0")

w, lse = bgd(ytr, xtr, nfeat, 10000, 0.1, 1.0e-6)
print "w_actual:", w_actual
print "w estimate", w
plt.plot(lse, label="Learning rate = 1.0e-6")

w, lse = bgd(ytr, xtr, nfeat, 10000, 0.1, 1.0e-5)
print "w_actual:", w_actual
print "w estimate", w
plt.plot(lse, label="Learning rate = 1.0e-5")

#w, lse = sgd(ytr, xtr, nfeat, 100000, 0.1, 1.0e-3)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3")

#w, lse = sgd(ytr, xtr, nfeat, 100000, 0.1, 1.0e-4)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-4")

#w, lse = adagrad(ytr, xtr, nfeat, 100000, 0.1, 1.0e-3)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3")

#w, lse = rmsprop(ytr, xtr, nfeat, 1000, 0.1, 1.0e-3, 0.1)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3, Forgetting factor = 0.1")

#w, lse = rmsprop(ytr, xtr, nfeat, 1000, 0.1, 1.0e-3, 0.01)
#print "w_actual:", w_actual
#print "w estimate", w
#plt.plot(lse, label="Learning rate = 1.0e-3, Forgetting factor = 0.01")

plt.xlabel("Number of iterations")
plt.ylabel("Loss function")
plt.title("Loss function for batch gradient descent method")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
